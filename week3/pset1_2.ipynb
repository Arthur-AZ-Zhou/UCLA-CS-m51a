## Background
The most common system for the symbolic representation of numbers in the world is the positional decimal numeral system. Invented between the 1st and 4th centuries by Indian mathematicians, it is based on ten symbols (0, 1, ... 9) and where the contribution of a digit at the $i^{th}$ position from the right to the value of a number is the value of the digit multiplied by $10^i$. The concept can be generalized to an arbitrary based b, such as the binary case where $b = 2$ and commonly used in computer systems. Other societies have used other bases, for example, the Babylonian numeral system used base 60 (it was the first positional system developed, while the Mayan numeral system was a vigesimal (base- 20) positional numeral system. There are many other examples of other bases, and there is a nice discussion at [this link](https://en.wikipedia.org/wiki/Positional_notation#Other_bases_in_human_language). Additionally there are number systems that are not even positional, e.g. Roman numeral system which uses additive and subtractive composition from the set of symbols I, V, X, L, C, D, and M, standing respectively for 1, 5, 10, 50, 100, 500, and 1,000 in our positional decimal numeral system. These have the disadvantage that the length of the strings representing a number grows rapidly.<br>Below we explore some unusual number systems used in computers.

## Negative Base Systems
In these systems there are $b$ digits with values $0$, $1$, ... $b-1$ and the base is $-b$. The value of the number $d_{n-1}\cdots d_1d_0$ (where $d_{n-1}>0$ for all $n>1$) is $\Sigma_{i=0}^{n-1}d_i(-b)^i$. Since even powers of negative numbers are positive while odd powers are negative, the digits in even positions get positive weights and those in odd positions get negative weight, thus allowing representation of both positive nad negative values without the use of the minus sign!<br>Consider the _negabinary_ system, which has base -2 and digits 0 and 1. ## Balanced Ternary System
One particularly interesting one is the *Balanced Ternary* system which uses three symbols $\{T,0,1\}$ where the numerical values of the three symbols are -1, 0, and 1 respectively, and the weight given to the digit (called _trit_) at the $i-{th}$ position (from the right, starting with $i=0$) is $3^i$. This number system allows expressing negative numbers without using a sign. For example, T01 will be $(-1) * 3^2 + 0 * 3^1 + 1 * 3^0 == -8$. In the early days of computing, a few experimental Soviet computers were built with balanced ternary instead of binary, and used three-valued ternary logic gates. Ternary logic can also be used for other purposes, such as logic statements that can take not only True and False values but also a value indicating Unknown.
